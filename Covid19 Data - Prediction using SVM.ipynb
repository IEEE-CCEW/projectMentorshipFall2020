# Prediction and Analysis using Support Vector Regression

```python
data = pd.read_csv('Covid19India.csv')
data
```

```python
data = data.iloc[:,1:]
data
```

```python
data.info()
```

```python
Day = data[data['Current Status']=='Hospitalized'].groupby(['Month','Day'])['Num Cases'].sum()
Day
```

```python
x = np.arange(len(Day))
x
```

```python
y = Day.values
y
```

```python
x = x.reshape(-1,1)
y = y.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.30,shuffle = True)
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
Sx = sc_X.fit_transform(X_train)
Sy = sc_y.fit_transform(y_train)
```

```python
print(Sx)
print("**********************************************************************************************")
print(Sy)
```

```python
from sklearn.svm import SVR
reg = SVR(kernel='rbf') #RBF => Radial Biased Function
print(reg.fit(Sx,Sy.ravel())) #ravel() converts 2d to 1d
```

```python
reg.score(Sx,Sy)*100
```
